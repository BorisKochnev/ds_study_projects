## Предобработка данных

### Устранение аномалий

Для замены пропусков сначала стоит проверить, насколько адекватны данные по самым дорогим квартирам:

```python
df.sort_values(by='last_price', ascending=False).head(10)
```

Данные выглядят правдопободно: у дорогих квартир внушительные площади, а количество комнат выше среднего. В следующих блоках изучение цены за квадратный метр позволит оценить, насколько такие данные правдоподобны

Перейду к изучению аномально дешёвых квартир:

```python
df.sort_values(by='last_price').head(5)
```

Аномальное значение одно, и кажется, что это техническая ошибка. От строки можно избавиться

```python
df.drop(labels = 8793,axis = 0, inplace = True)
```

Теперь стоит обратить внимание на необоснованно высокие потолки, насколько такие значения влияют на среднее значение:

```python
df[['total_area','ceiling_height', 'last_price']].sort_values(by='ceiling_height', ascending=False).head(10)
```

За исключением одной квартиры с потолками 100 м, можно предположить, что некоторые данные введены с ошибкой. Если разделить эти данные на 10, данные будут похожи на правду: потолки 2.5 м, 2.7 м, 3.2 м

Теперь стоит оценить квартиры с аномально низкими потолками:

```python
df[['total_area','ceiling_height', 'last_price']].sort_values(by='ceiling_height').head(15)
```

Данные о потолках 1 и 2 м выглядят неправдоподобно, однако можно сделать предположение, что квартиры с потолками 2 м существуют. Изучив размах выборки, можно приступать к замене аномальных значений:
* От строк с потолками 1-1.75 м и 100 м можно избавиться
* Строки с перепутанным разрядом можно поделить на 10. Предполагаю, что минимально возможная высота потолков - 2 метра. Значит в условие попадут все квартиры, чья высота больше или равна 20 метрам

```python
df.drop(labels = [22590, 5712, 16934, 22869],axis = 0, inplace = True)
df.loc[df['ceiling_height'] >= 20, 'ceiling_height'] = df['ceiling_height'] / 10
df[['total_area','ceiling_height', 'last_price']].sort_values(by='ceiling_height', ascending=False).head(20)
```

В данных остались небольшие аномалии с потолками высотой 8-14 метров, однако такие значения не повлияют на общую выборку. Вернусь к ним на следующем этапе при более подробном изучении столбцов

В выборке есть данные с нулевым количеством комнат. Проверю, связано ли это с тем, является ли квартира студией:

```python
df.pivot_table(index='rooms',values='studio', aggfunc=('count', 'sum', 'mean'))
```

Да, связано, причем в некоторых случаях студии считаются квартирами с одной комнатой. При этом 30% квартир с нулевым количеством комнат не являются студиями. Есть два варианта решения этого вопроса:
* считать, что только студии могут иметь нулевое количество комнат. В этом случае можно сравнить медианный метраж студий и "не студий" с нулевым количеством комнат. Возможно, некоторые квартиры с нулевым количеством комнат ошибочно не определены как студии
* считать, что все квартиры могут иметь минимум 1 комнату, ведь технически студия - это однокомнатная квартира без перегородок

Кажется, что наиболее наглядным решением для дальнейшего анализа будет считать студию квартирой с нулевым количеством комнат. В этом случае мы получим удобную группировку данных, в которой студии и все остальные квартиры сгруппированы по комнатам. Но стоит проверить рациональность этой гипотезы:

```python
zero_rooms_studio = df.query('rooms == 0 and studio == True')['total_area'].median()
one_room_studio = df.query('rooms == 1 and studio == True')['total_area'].median()
zero_rooms_flat = df.query('rooms == 0 and studio == False')['total_area'].median()
one_room_flat = df.query('rooms == 1 and studio == False')['total_area'].median()

studio_room_comparision = {'студия': [zero_rooms_studio, one_room_studio]
                        , 'квартира': [zero_rooms_flat, one_room_flat]
                          }

pd.DataFrame(data=studio_room_comparision)
```

Из данных можно сделать несколько выводов:
* Медианная площадь студий с одной комнатой и без комнаты почти не различаются
* Медианная площадь однокомнатной квартиры выше, чем квартиры-нестудии с нулевым количеством комнат

Самым логичным решением будет назначить всем квартирам с нулевым количеством комнат статус "студия", а для студий проставить нулевые значения в столбце `rooms` 

Такое решение может быть логичным, так как теперь студии можно будет определить и по количеству комнат, что выглядит логичным: чем больше количество комнат, тем больше площадь, а самая маленькая площадь у студии, где отдельной комнаты нет

```python
df.loc[df['rooms'] == 0, 'studio'] = True
df.loc[df['studio'] == True, 'rooms'] = 0
df.pivot_table(index='rooms',values='studio', aggfunc='sum')
```

Теперь это студии - это квартиры с нулевым количеством комнат 

### Замена пропусков

Сначала заполню значения null в тех столбцах, где подразумевается 0:
* `is_apartment` - предполагаем, что значения null означают 0
* `balcony` - предполагаем, что значения null означают 0
* `ponds_around3000`  - предполагаем, что значения null означают 0
* `parks_around3000` - предполагаем, что значения null означают 0

```python
zero_null_columns = ['is_apartment', 'balcony', 'ponds_around3000', 'parks_around3000']
df[zero_null_columns] = df[zero_null_columns].fillna(0)
```

Затем избавлюсь от тех значений, заменить которые не представляется возможным:
* `floors_total` - значений немного, можно избавиться от данных с пропущенными значениями
* `locality_name` - определить название населенного пункта в датасете не представляется возможным. С учетом небольшого количества значений от этих данных можно отказаться
* `days_exposition` - определить время активности объявления не представляется возможным

```python
df.dropna(subset=['floors_total', 'days_exposition'], inplace=True)
df['locality_name'] = df['locality_name'].fillna('unknown')
df.shape
```

Затем заменю неизвестные значения на медианные:
* `living_area` - определить жилую площадь квартиры можно по общей площади и медианным значениям в зависимости от количества комнат
* `ceiling_height` - предполагается, что размах высоты потолков не высок, подойдет замена на медианные значения
* `kitchen_area` - определить жилую площадь квартиры можно по общей площади и медианным значениям в зависимости от количества комнат

```python
#Вычисляю долю жилой площади от общей площади для каждой квартиры 
df['living_share'] = (df['living_area'] / df['total_area'])
#Значения null в столбце 'living_share' меняю на медианные значения в зависимости от количества комнат 
df['living_share'] = df['living_share'].fillna(df.groupby('rooms')['living_share'].transform('median'))
#Через долю жилой площади вычисляю абсолютное значения жилой площади для неизвестных значений
df['living_area'] = df['living_area'].fillna(df['total_area'] * df['living_share'])

#Повторяю процедуру для кухонь
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
df['kitchen_share'] = (df['kitchen_area'] / df['total_area'])
df['kitchen_share'] = df['kitchen_share'].fillna(df.groupby('rooms')['kitchen_share'].transform('median'))
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area'] * df['kitchen_share'])

#Чтобы соблюдать порядок выполнения задания, избавлюсь от лишних столбцов, добавлю их позже
df = df.drop(['kitchen_share','living_share'], axis=1)
df.info()
```

Заменены все планируемые значения null за исключением неоторые значений `kitchen_area`. Возможно, дело опять касается квартир-студий:

```python
df[df['kitchen_area'].isna()].groupby('rooms')['rooms'].count()
```

Да, действительно так. В студиях кухни, как правило, находятся на жилой площади, поэтому эти значения можно смело заменить на 0:

```python
df['kitchen_area'] = df['kitchen_area'].fillna(0)
df.info()
```

Итого я избавился от всех значений null за исключением `airports_nearest`, `cityCenters_nearest`, `parks_nearest`, `ponds_nearest` - эти значения оставляю, так как они означают, что рядом с квартирой нет соответствующего объекта

## Расчёты и добавление результатов в таблицу

В рамках задания нужно добавить новые столбцы в таблицу:
* `square_meter_price` - цена квадратного метра;
* `weekday` - день недели публикации объявления;
* `month` - месяц публикации объявления;
* `year` - год публикации объявления;
* `floor_category` - категория этажа (первый, последний, другой);
* `living_share` - соотношение жилой и общей площади
* `kitchen_share` - соотношение кухни и общей площади

```python
# Добавляю цену квадратного метра:
df['square_meter_price'] = (df['last_price'] / df['total_area']).astype('int')
# Добавляю день недели публикации объявления:
df['weekday'] = df['first_day_exposition'].dt.weekday
# Добавляю месяц публикации объявления:
df['month'] = df['first_day_exposition'].dt.month
# Добавляю год публикации объявления:
df['year'] = df['first_day_exposition'].dt.year
# Добавляю соотношение жилой и общей площади:
df['living_share'] = df['living_area'] / df['total_area']
# Добавляю соотношение кухни и общей площади:
df['living_share'] = df['kitchen_area'] / df['total_area']
```

Чтобы категоризировать этажность, стоит сделать функцию, принимающую на вход строку. Функция будет проверять, какой категории принадлежит этаж:

```python
def floor_categorizer(row):
    if row['floor'] == 1:
        return 'first'
    if row['floor'] == row['floors_total']:
        return 'last'
    return 'other'
```

Далее применю функцию для нового столбца:

```python
df['floor_category'] = df.apply(floor_categorizer, axis=1)
```

Чтобы посчитать корреляцию цены и категории этажа, нужно категории этажа назначить числа, например 0,1,2. Для этого подойдет функция. Однако перед этим стоит понять, как различаются цены по разным категориям этажа:

```python
df.pivot_table(index='floor_category', values ='last_price', aggfunc='median').astype('int')
```

```python
def floor_categorizer_int(row):
    if row['floor_category'] == 'first':
        return 0
    if row['floor_category'] == 'last':
        return 1
    return 2
```

```python
df['floor_category_int'] = df.apply(floor_categorizer_int, axis=1)
df[['last_price', 'floor_category_int']].corr()
```

Между категорией этажа и ценой не выявлено линейной корреляции. Однако по медианным значениям можно увидеть, что стоимости квартир на первом, последнем и других этажах отличаются. Скорее всего, зависимость цены от категории этажа не описывается линейной корреляцией
